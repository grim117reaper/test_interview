```python
#imports
import pandas as pd
```

```python
df_amazon = pd.read_csv("raw_data_amazon.csv")
df_md = pd.read_csv("raw_data_md.csv")
```

```python
df_amazon
```

```python
df_md
```

```python
combined = pd.concat([df_amazon,df_md])
```

```python
combined
```

```python
gst_rates = pd.read_csv("gst_rates.csv")
```

```python
gst_rates["gst_rate"]=gst_rates["gst_rate"].str[:-1].astype(int)
```

```python
gst_rates.set_index("category",inplace=True)
```

```python
gst_rates_dict = gst_rates.to_dict()["gst_rate"]
```

```python
gst_rates_dict
```

```python
combined["gst_sub"] = 0
```

```python
for idx,row in combined.iterrows():
    combined.loc[idx,"gst_sub"] = row["price"] * 100 / (gst_rates_dict[row["category"]] + 100)
```

```python
combined
```

```python
amazon_avg_total = combined.loc[combined["source"] == "amazon"]["gst_sub"].mean()
md_avg = combined.loc[combined["source"] == "mdcomputers"]["gst_sub"].mean()
```

```python
categories = combined.category.unique()
```

```python
amazon_avg = {}
md_avg = {}
amazon = combined.loc[combined["source"]=="amazon"]
md = combined.loc[combined["source"]=="mdcomputers"]
for x in categories:
    amazon_avg[x] = amazon.loc[amazon["category"] == x]["gst_sub"].mean()
    md_avg[x] = md.loc[md["category"] == x]["gst_sub"].mean()
```
